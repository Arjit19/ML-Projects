```python
def Cleaning_Data(df):
  df.dropna(inplace=True)
  df["Entry DateTime"]= pd.to_datetime(df["Entry DateTime"])
  df["Profit/Loss (C)"]=df["Profit/Loss (C)"].apply(pd.to_numeric)
  df["Total Efficiency"]=df["Total Efficiency"].apply(lambda x:pd.to_numeric(x[0:len(x)-1]))
  df["Exit Efficiency"]=df["Exit Efficiency"].apply(lambda x:pd.to_numeric(x[0:len(x)-1]))
  df["Duration"]=pd.to_timedelta(df["Duration"])
  df["Duration"]=df["Duration"].apply(lambda x:x.seconds)
  df.drop(["Symbol","Trade Type","Exit DateTime","Entry Price","Exit Price",'Trade Quantity', 'Max Open Quantity',
       'Max Closed Quantity','Cumulative Profit/Loss (C)','High Price While Open', 'Low Price While Open','Entry Efficiency','FlatToFlat Profit/Loss (C)',
       'FlatToFlat Max Open Profit (C)', 'FlatToFlat Max Open Loss (C)','Note','Commission (C)', 'Open Position Quantity', 'Close Position Quantity'],axis=1,inplace=True)
  df["Hour"]=df["Entry DateTime"].apply(lambda x:x.hour)
  df["Month"]=df["Entry DateTime"].apply(lambda x:x.month)
  df["Year"]=df["Entry DateTime"].apply(lambda x:x.year)
  df.drop("Entry DateTime",axis=1,inplace=True)
  df_f=df.groupby(["Year","Month","Hour"]).mean()
  df_f["Day"]=1
  df_f.reset_index(inplace=True)
  df_f["YearMonth-Hour"]=pd.to_datetime(df_f[["Year","Month","Day","Hour"]])
  df_f.drop(["Year","Month","Day","Hour"],axis=1,inplace=True)
  df_f.set_index("YearMonth-Hour",inplace=True)
  df_f=df_f.resample("H").mean()
  df_f.reset_index(inplace=True)
  df_f["Day"]=df_f["YearMonth-Hour"].apply(lambda x:x.day)
  for x in range(len(df_f)):
    if df_f["Day"][x] !=1 :
        df_f.drop([x],inplace=True)
  df_f.set_index("YearMonth-Hour",inplace=True)
  df_f.drop("Day",axis=1,inplace=True)
  return df_f


```

```python
def Fillwith_RollingMean(d):
  df=d.copy()
  df["Profit/Loss (C)"].fillna(df["Profit/Loss (C)"].rolling(49,center=True,min_periods=1).mean(),inplace=True)
  df["Duration"].fillna(df["Duration"].rolling(49,center=True,min_periods=1).mean(),inplace=True)
  df['Exit Efficiency'].fillna(df['Exit Efficiency'].rolling(49,center=True,min_periods=1).mean(),inplace=True)
  df['Max Open Profit (C)'].fillna(df['Max Open Profit (C)'].rolling(49,center=True,min_periods=1).mean(),inplace=True)
  df['Max Open Loss (C)'].fillna(df['Max Open Loss (C)'].rolling(49,center=True,min_periods=1).mean(),inplace=True)
  df["Total Efficiency"].fillna(df["Total Efficiency"].rolling(49,center=True,min_periods=1).mean(),inplace=True)
  return df
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def backtesting_withRF(d):
  df=d.copy()
  df["Profit Shifted"]=df["Profit/Loss (C)"].shift(-24)
  df.dropna(inplace=True)
  x = np.array(df[['Profit/Loss (C)', 'Duration', 'Exit Efficiency', 'Max Open Profit (C)','Max Open Loss (C)', 'Total Efficiency']])
  y = np.array(df['Profit Shifted']).reshape(-1,1)

  n= 24*24
  predictions = []
  true_values = []
  for i in range(n,len(x),24):
    x_train = x[0:i]
    y_train = y[0:i]
    x_test = x[i:i+24]
    y_test = y[i:i+24]
    model = RandomForestRegressor(n_estimators=100,verbose=2)
    model.fit(x_train, y_train)
    predictions.append(model.predict(x_test))
    true_values.append(y_test)
        
  pred=np.array(predictions).reshape(len(predictions)*24,1)[:,-1]
  real=y[24*24:,:][:,-1]
  mae=mean_absolute_error(real,pred)
  rms=np.sqrt(mean_squared_error(real,pred))
  return mae,rms
```

```python
simulations=[sim100_final,sim101_final,sim102_final,sim103_final,sim110_final,sim111_final,sim112_final,sim113_final,sim120_final,sim121_final,sim122_final,sim123_final,sim130_final,sim131_final,sim132_final,sim133_final]
mae={}
rms={}
i=0
for s in simulations:
  mae[i],rms[i]=backtesting_withRF(s)
  i=i+1




```

```python
def Final_Predictions(d):
  df=d.copy()
  x_test = np.array(df[['Profit/Loss (C)', 'Duration', 'Exit Efficiency', 'Max Open Profit (C)','Max Open Loss (C)', 'Total Efficiency']])
  x_test = x_test[len(x_test)-24 : len(x_test)]
  df["Profit Shifted"]=df["Profit/Loss (C)"].shift(-24)
  df.dropna(inplace=True)
  x = np.array(df[['Profit/Loss (C)', 'Duration', 'Exit Efficiency', 'Max Open Profit (C)','Max Open Loss (C)', 'Total Efficiency']])
  y = np.array(df['Profit Shifted']).reshape(-1,1)

  
  x_train = x
  y_train = y
  
  model = RandomForestRegressor(n_estimators=100,verbose=2)
  model.fit(x_train, y_train)
  future_predictions= model.predict(x_test)
  return future_predictions
```

```python
Comparing_Bots=pd.DataFrame(index=range(0,24),columns=["sim100_final","sim101_final","sim102_final","sim103_final","sim110_final","sim111_final","sim112_final","sim113_final","sim120_final","sim121_final","sim122_final","sim123_final","sim130_final","sim131_final","sim132_final","sim133_final"])
```
